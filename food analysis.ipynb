Creating different age groups for the data

```python
bins=[10,20,30,40,50]
labels=['10-20','20-30','30-40','40-50']
df['age_group']=pd.cut(
    df['age'],bins=bins,labels=labels,include_lowest=True
)
```

```python
df
```

```python
df.insert(1,'age_group',df.pop('age_group'))
```

```python
df
```

finding which age group spent the most amount of  money in which city

```python
data=df.pivot_table(index='age_group',columns='city',values='order_value',aggfunc='sum',observed=False)
```

```python
data
```

```python
sns.heatmap(data,cmap='viridis')
```

it is age group 20-30 that spent the most money ordering online food in all the cities in the data

plot of order frequency among different age groups

```python
temp=df['age_group'].value_counts().sort_values()
sns.countplot(data=df,
              x='age_group',
              order=temp.index,hue='age_group')
plt.title("Order Frequency Among Different Age Groups")
plt.xlabel("age groups")
plt.ylabel("order frequency")
plt.show()
```
